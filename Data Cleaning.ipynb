```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df =pd.read_csv('Dirty Data.csv')
df.head()
```

```python
df.rename(columns = {'Bio data with information on name, Email, age, City of Birth, and Gender': 'unname'},inplace =True)
df.head()
               
```

```python
df['Name']= df['unname'].apply(lambda x: x.split('Email',1)[0] if 'Email' in x else x)
df
```

```python
df['email'] =df['unname'].apply(lambda x: x.split('Email',1)[1].split('@@',1)[0] 
             + '@' + x.split('@@',1)[1].split(';',1)[0] if 'Email' in x else x)
df.head()
```

```python
df['Email']=df['email'].str.extract(r'([^\d]+)\d+')
df.head()
```

```python
df['Age'] = df['email'].apply(lambda x: x.split('.com',1)[1])
df.head()

                              
```

```python
df['code'] = df['unname'].apply(lambda x:x.split('Years',1)[1])
df['Name']=df['Name'].str.replace('-' ,' ')	
df.head()
                                
```

```python
df['City of Birth']=df['code'].apply(lambda x:x.split('Gender',1)[0])
df['Gender']=df['code'].apply(lambda x:x.split('Gender',1)[1])
data =df[['Name','Email','Age','City of Birth','Gender']]
data


                            
```

```python
data.head()
```

```python
 g1 = data['Gender'].value_counts().sort_index()
g1
```

```python
plt.figure(figsize=(7,4))
g1.plot(kind = 'bar',color =('#768E36','#52A6F7'))
plt.xlabel('gender',fontsize= 15,color ='black')
plt.ylabel('counts',fontsize =15,color ='black')
plt.title('Total male and female', fontweight= 'bold')
plt.grid(axis='y',linestyle ='-',alpha= 0.3,color = '#0F0F0F')
plt.xticks(rotation=45)
plt.show() 
```

```python
g3 =data[data['Age'].astype(int)> 20]
g3
```

```python
g2 =g3['Gender'].value_counts()
g2
```

```python
plt.figure(figsize=(7,4))
g2.plot(kind='pie',color=('#5FD8D4','#E45F8E'), autopct=lambda x: f'{x:.1f}% ({int(x*sum(g2)/100)})')
plt.title('Male and Female above 20')
plt.ylabel('')
plt.show()
        
```
