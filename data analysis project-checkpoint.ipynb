```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('project2.csv')
df.head(10)
```

```python
df.info()
```

```python
# removing the characters
df['Company Name']=df['Company Name'].str.split('\n',expand=True)[0]
df.head()
```

```python
# replacing all the value with meaningful way
df['Size']= df['Size'].str.replace('to','-')
df['Revenue']= df['Revenue'].str.replace('to','-')
df['Competitors']= df['Competitors'].str.replace('-1','not found yet')
df.head(5)
```

```python
# replacing all -1 value from any columns to 'unknown'
df.replace(-1,'unknown',inplace =True)
df.head()
```

```python
df.isna().sum()
```

```python
# Find the Data Scientist position job that located in new york
data = df[df['Job Title'].str.contains(r'Data Scientist')]
data= data[data['Headquarters'].str.contains(r'New York, NY')]
data = data[['Job Title','Headquarters']]
data.head()
```
